```python
# Carrego o arquivo com o resultado da eleição por município
df_votacao = pd.read_csv('./dados_eleicao/votacao_candidato_munzona_2018_BR.csv', sep=';', encoding='latin1')

df_votacao.head()
```

```python
# Visualizar o dataframe transposto
df_votacao.head().T
```

```python
# Separar os dados, pois precisamos apenas do 2º turno
df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]

df_votacao.sample(3).T
```

```python
# Vamos verificar quantos municípios unicos existem no conjunto de dados
# vamos pegar por código do município
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Mas o brasil tem 5570 municípios
# O arquivo traz os munícpios do exterior que possuem votação]
# vamos verificar algumas cidades fora do brasil
df_votacao[df_votacao['SG_UF'] == 'ZZ'].sample(4).T
```

```python
# Quantas cidades fora do brasil estão no conjunto de dados
df_votacao[df_votacao['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# Vamos eliminar os municípios de fora do Brasil
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```python
# Verificando a quantidade de municípios
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separados por município/zona
# Municípios pequenos normalmente tem apenas 1 zona eleitora
# Mas municípios maiores podem ter várias zonas eleitorais

# Vamos verficiar quantas zonas eleitorais tem em Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE' ]
```

# Devemos agrupar os dados por municipio
# Serão 2 linhas para cada município, uma para cada candidato

```python
# Primeiro vamos eliminar algumas colunas do dataframe que não serão utilizadas
df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 
                         'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

df_votacao.head()
```

```python
# Precisamos agregar os dados por município e candidato
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# para ajustar o dataframe que ficou todo agrupado
# devemos fazer reindexação
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()
```

```python
df_votacao.sample(4)
```

```python
# Vamos verficiar novamente a cidade de belo horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```python
# Ordenamos o dataframe pela quantidade de votos nominais e depois deixar apenas o primeiro registro de cada município
# Vejo como fincaria a ordenação
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head()
```

```python
# Agora sim ordenamos
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

#dropamos as linha sduplicadas, mantendo a primeira (será mantida nesse caso a que tem mais votos, pois ordenamos do maior para o menor)
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head(10)
```

```python
# Criar coluna representando o resultado da eleição naquele município
# Sendo True se o candidato venceu e False se o candidato Perdeu
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
df_votacao.head(10)
```

```python
# Mostrando alguns valores do dataframe do geopandas
df_mapa.sample(3)
```

```python
# Tentando encontrar município do df_votacao atraves do código do município do df_mapa
# Se não achar nada é porque os códigos são diferentes
df_votacao[df_votacao['CD_MUNICIPIO'] == 3300308]
```

```python
# Porque não podemos usar o nome do município para juntar os dados?
# Pois temos várias cidades no brasil com o mesmo nome, como por exemplo 'OUTRO BRANCO'
df_votacao[df_votacao['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
# Recuperar a tabela de correspondência entre IBGE e TCE do github do ESTADAO:
url = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(url)

df_equivalencia.head()
```

# Prepara os dataframes

```python
# Vamos reindexar o dataframe de votação pelo código do TSE
# Crio um dataframe novo
df_vot_novo = df_votacao.copy()

# altero o index de cada item para o código do município.
# no momento de fazer um join (juntar) esse dataframe com o outro, se os itens no qual queremos juntar
# tiverem o mesmo indice, o processo fica mais fácil
# o drop indica que vamos manter a coluna cd_município além de transformá-la em indixe
df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)
```

```python
df_vot_novo.head()
```

```python
# Vamos também reindexar o dataframe de equivalencia pelo código do TSE
df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```python
# Agora que os dois dataframes estão indexados pelo mesmo código, podemos uní-los
# antes vamos verificar se os dois possuem o mesmo tamanho
df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
# Juntando os dataframes em um novo dataframe
df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```python
# Vamos juntar o dataframe do geopandas com o dataframe votação/equivalencia
# primeiro verificamos os tamanhos
df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# Verificamos dois problemas:
# o dataframe do mapa possui 2 registros a mais
# e o código do ibge do dataframe votação/equivalência, está como INTEIRO e o do mapa está como OBJECT

# Vamos transformar a coluna de código do IBGE do dataframe de Votação Equivalencia
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
# Conversão feita 
# Vamos reindexar os indices para o código do IBGE

# Mudando o indice do df_mapa
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)
```

```python
# Mudando o indice do dataframe de votação/equivalência
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
# Falta verificar a diferença do tamanho nos dataframes
# mas podemos juntar os dataframes e ver quais os itens que ficam com colunas vazias.
# esses itens serão essa diferença de tamanho
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head()
```

```python
# Verificando o tamanho do dataframe mapa_novo
df_mapa_novo.info()
```

```python
# Verificamos com o info que algumas colunas possuem valores nullos
# Vamos descobrir os 2 municípios a mais no mapa
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
# Remover a coluna de AJUSTE, pois está nula
df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
# Esses itens listados, LAGOA MIRIM e LAGOA DOS PATOS não são munícipios 
# Sendo assim podemos excluí-los do datafram
# DROPNA deleta itens do dataframe que possuem valores nulos
df_mapa_novo.dropna(inplace=True)
```

```python
df_mapa_novo.info() 
```

```python
# Vamos desenhar o mapa

# Criar a referência para o código de cor (False = red, TRue = green)
from matplotlib.colors import ListedColormap

# Criamos uma lista de cores
cmap  = ListedColormap(['red', 'green'])

# Desenhando o mapa
# Definimos o tamanho do mapa co mo tamanho informado
fig, ax = plt.subplots(1, figsize=(16,16))
# Pegamos o dato que será usado no gráfico com o parametro column
# cmap é a lista de cores
# legend informamos se o mapa terá legenda
# linewidth é a espessura da linha que separa os municípios
# edgecolor é a cor da linha que separa os municípios
ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.1, ax=ax,
                      edgecolor='white')

# Esconder os eixos x e y
ax.set_axis_off()

# Colocando título do mapa
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize':'25', 'fontweight':'3'})

# Criar uma nota de rodapé
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019',
        xy=(0.1, 0.08), xycoords='figure fraction', ha='left',
        va='top', fontsize=10, color='#555555')

# ALterar a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad') #FALSE
leg.get_texts()[1].set_text('Jair Bolsonaro') #TRUE

plt.show()
```
