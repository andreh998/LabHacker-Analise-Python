# Análise do eleitorado brasileiro

```python
# Importando as bibliotecas
import pandas as pd
```

```python
# Carregar o arquivo
df = pd.read_csv('eleitorado_municipio_2018.csv', sep = ';', encoding = 'latin1')

df.head()
```

```python
# Mudamos a visualização para colunas
df.head().T
```

```python
# Tipos e quantidades
df.info()
```

```python
# As 5 cidades com maior quantidade de eleitores deficientes
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras: ', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores: ', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não informado: ', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criar variaveis para calcular o percentual
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_feminino = df['QTD_ELEITORES_FEMININO'].sum()
tot_masculino =  df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()

# Mostrar percentuais
print('Eleitoras: ', ((tot_feminino/tot_eleitores) * 100).round(2))
print('Eleitores: ', ((tot_masculino/tot_eleitores) * 100).round(2))
print('Não informado: ', ((tot_nao/tot_eleitores) * 100).round(2))
```

```python
# Quantas cidades com mais homens que mulheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Vamos criar uma nova coluna para indicar a relação fem/masc
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais são os municipios com maior relação f/m
df.nlargest(5, 'RELACAO_FM')
```

```python
# Quais os municípios com a menor relação f/m
df.nsmallest(5, 'RELACAO_FM').T
```

```python
# vamos criar um novo dataframe só com os municípios do Brasil
# O copy mantem os dados do dataframe original e cria um novo dataframe
# Sem usar o copy, qualquer alteração feita tanto no df antigo quando no df novo, ele altera em ambos os dfs
df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head()
```

```python
# Quais são os municipios com maior relação f/m
df_brasil.nlargest(5, 'RELACAO_FM').T
```

```python
# Quais os municípios com a menor relação f/m
df_brasil.nsmallest(5, 'RELACAO_FM').T
```

```python
# plotar um gráfico de distribuição da relação f/m
%matplotlib inline

df_brasil['RELACAO_FM'].plot.hist(bins=100)
```

```python
# Importando as bibliotecas gráficas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotar a distribuição da relação f/m
sns.distplot(df_brasil['RELACAO_FM'], bins=100, color='red', kde=False)

# embelezando o gráfico
plt.title('Relação eleitoras/eleitores', fontsize = 18)
plt.xlabel('Eleitoras/Eleitores', fontsize = 14)
plt.ylabel('Frequência', fontsize = 14)
plt.axvline(1.0, color='black', linestyle = '--')
```

```python
# verificar a distribuição f/m mas mostrando os pontos (municípios)
sns.swarmplot(data = df_brasil, x = 'NM_REGIAO', y = 'RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='--')
```

```python
# plotar a quantidade total de eleitores por faixa etária
# usando o gráfico de barras horinzontal

# primeiro vamos Listar as colunas de nosso interesse
lista = ['QTD_ELEITORES_16', 'QTD_ELEITORES_17', 'QTD_ELEITORES_18',
        'QTD_ELEITORES_19', 'QTD_ELEITORES_20']

tot_idade = df[lista].sum()
```
